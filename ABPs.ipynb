```python
import pandas as pd
import hvplot.pandas
import holoviews as hv
import matplotlib.pyplot as plt
from bokeh.io import export_png
import glob
import os
```

CARGAR TODOS LOS ARCHIVOS Y CONCATENARLOS TODOS

```python
#Hacer una lista con los archivos de la carpeta
archivos = glob.glob("/Users/nacho/Desktop/TFM/Datos/Datos equipos LaLiga J27/*xlsx")
len(archivos)
```

```python
#Cargar los archivos
len(archivos)

lista_de_archivos = []
for f in archivos:
    data = pd.read_excel(f)
    nombre = os.path.basename(f)
    data['carpeta_contenedora'] = nombre
    lista_de_archivos.append(data)
```

```python
len(lista_de_archivos)
```

```python
#Concatenarlos en un solo data frame
df = pd.concat(lista_de_archivos, ignore_index= True)
```

```python
df.shape
```

LIMPIAR Y ORDENAR LOS DATOS DEL ARCHIVO CONCATENADO

```python
#Eliminar filas de cada partido, solo nos quedamos con las generales, las estadísticas de lo equipos y las de los rivales contra cada equipo.
df = df[~(df['Team'].str.startswith('2022') | df['Team'].str.startswith('2023'))]
```

```python
#Crear dos df distientos uno con las estadísticas de lo equipos y otros con las estadísticas de los rivales contra cada equipo.
Opponents_stats = df[df['Team'].str.startswith('Opponents')]
Team_stats = df[~df['Team'].str.startswith('Opponents')]
```

4 ABPs

```python
#Calcular las medias de las columnas de centros y pases profundos.
corners_mean = Team_stats['Corners'].mean()
corners_finalizacion_mean = Team_stats['% of corners with shots'].mean()

#Crear el gráfico con los colores específicos para cada equipo
corners_rivales = Team_stats.hvplot.scatter(x='Corners', y='% of corners with shots', by='Team', 
                                 legend='right', height=600, width=900, color='color', 
                                 hover_cols=["Corners", "% of corners with shots"])

#Añadir lineas de media de goles y tiros para ver cuadrantes
corners_mean_line = hv.VLine(corners_mean)
corners_mean_line.opts(
    color='grey', 
    line_dash='dashed', 
    line_width=1,
)

corners_finalizacion_mean_line = hv.HLine(corners_finalizacion_mean)
corners_finalizacion_mean_line.opts(
    color='grey', 
    line_dash='dashed', 
    line_width=1,
)

corners_rivales = corners_rivales * corners_mean_line * corners_finalizacion_mean_line.opts(title='Corners por 90')

corners_rivales
```

```python
#Calcular las medias de las columnas de centros y pases profundos.
set_pieces_mean = Team_stats['Set pieces'].mean()
set_pieces_finalizacion_mean = Team_stats['% of set pieces with shots'].mean()

#Crear el gráfico con los colores específicos para cada equipo
set_pieces = Team_stats.hvplot.scatter(x='Set pieces', y='% of set pieces with shots', by='Team', 
                                 legend='right', height=600, width=900, color='color', 
                                 hover_cols=["Set pieces", "% of set pieces with shots"])

#Añadir lineas de media de goles y tiros para ver cuadrantes
set_pieces_mean_line = hv.VLine(set_pieces_mean)
set_pieces_mean_line.opts(
    color='grey', 
    line_dash='dashed', 
    line_width=1,
)

set_pieces_finalizacion_mean_line = hv.HLine(set_pieces_finalizacion_mean)
set_pieces_finalizacion_mean_line.opts(
    color='grey', 
    line_dash='dashed', 
    line_width=1,
)

set_pieces = set_pieces * set_pieces_mean_line * set_pieces_finalizacion_mean_line.opts(title='Jugadas a balón parado por 90')

set_pieces
```

```python
#Calcular las medias de las columnas de centros y pases profundos.
free_kicks_mean = Team_stats['Free kicks'].mean()
free_kicks_finalizacion_mean = Team_stats['% of free kicks with shots'].mean()

#Crear el gráfico con los colores específicos para cada equipo
free_kicks = Team_stats.hvplot.scatter(x='Free kicks', y='% of free kicks with shots', by='Team', 
                                 legend='right', height=600, width=900, color='color', 
                                 hover_cols=["Free kicks", "% of free kicks with shots"])

#Añadir lineas de media de goles y tiros para ver cuadrantes
free_kicks_mean_line = hv.VLine(free_kicks_mean)
free_kicks_mean_line.opts(
    color='grey', 
    line_dash='dashed', 
    line_width=1,
)

free_kicks_finalizacion_mean_line = hv.HLine(free_kicks_finalizacion_mean)
free_kicks_finalizacion_mean_line.opts(
    color='grey', 
    line_dash='dashed', 
    line_width=1,
)

free_kicks = free_kicks * free_kicks_mean_line * free_kicks_finalizacion_mean_line.opts(title='Jugadas a balón parado por 90')

free_kicks
```

```python
#Calcular las medias de las columnas de centros y pases profundos.
penalties_mean = Team_stats['Penalties'].mean()
penalties_converted_mean = Team_stats['Penalties converted'].mean()

#Crear el gráfico con los colores específicos para cada equipo
penalties = Team_stats.hvplot.scatter(x='Penalties', y='Penalties converted', by='Team', 
                                 legend='right', height=600, width=900, color='color', 
                                 hover_cols=["Penalties", "Penalties converted"])

#Añadir lineas de media de goles y tiros para ver cuadrantes
penalties_mean_line = hv.VLine(penalties_mean)
penalties_mean_line.opts(
    color='grey', 
    line_dash='dashed', 
    line_width=1,
)

penalties_converted_mean_line = hv.HLine(penalties_converted_mean)
penalties_converted_mean_line.opts(
    color='grey', 
    line_dash='dashed', 
    line_width=1,
)

penalties = penalties * penalties_mean_line * penalties_converted_mean_line.opts(title='Penatis por 90')

penalties
```
