6.1Porteros

```python
# Define las métricas y sus respectivos pesos
metricas = ["Pases largos exitosos x90 (número)", 
            "Paradas, %", 
            "Pases progresivos exitosos x90 (número)", 
            "Tiros por gol en contra",
            "Goles prevenidos x90",
            'Duelos aéreos del portero x90']

pesos = [0.25, 0.5, 0.25, 0.5, 0.5, 0.50]

# Calcula la media y la desviación estándar de cada métrica
means = porteros_df[metricas].mean()
stds = porteros_df[metricas].std()

# Normaliza los datos utilizando la normalización por z-score
normalized_data = (porteros_df[metricas] - means) / stds

# Calcula la puntuación por estadística multiplicada por su peso
for i, metrica in enumerate(metricas):
    porteros_df[metrica + '_puntuacion'] = normalized_data[metrica] * pesos[i]

# Calcula la puntuación total sumando las puntuaciones ponderadas por cada estadística
porteros_df['puntuacion_total'] = porteros_df[[metrica + '_puntuacion' for metrica in metricas]].sum(axis=1)

# Normaliza las puntuaciones totales entre 0 y 10
puntuacion_maxima = porteros_df['puntuacion_total'].max()
puntuacion_minima = porteros_df['puntuacion_total'].min()

porteros_df['Rendimiento'] = ((porteros_df['puntuacion_total'] - puntuacion_minima) / (puntuacion_maxima - puntuacion_minima)) * 100

# Convierte la puntuación a enteros
porteros_df['Rendimiento'] = porteros_df['Rendimiento'].astype(int)
```
